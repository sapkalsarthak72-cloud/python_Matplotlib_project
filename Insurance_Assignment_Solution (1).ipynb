# Insurance Dataset EDA Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Download the dataset if it's not already present


df=pd.read_csv('insurance.csv')
df.head(29)
```

```python
# Q2
print(df.shape)
df.info()
display(df.describe())
```

```python
# Q3
print(df.isnull().sum())
```

```python
# Q4
num=df.select_dtypes(include='number').columns
cat=df.select_dtypes(exclude='number').columns
print(num)
print(cat)
```

```python
# Q5
for c in num:
    plt.figure(figsize=(5,3))
    sns.histplot(df[c],kde=True)
    plt.title(f'Distribution of {c}')
    plt.show()
```

```python
# Q6
for c in cat:
    plt.figure(figsize=(5,3))
    sns.countplot(data=df,x=c)
    plt.title(f'Count of {c}')

    plt.show()
```

```python
# Q7
plt.figure(figsize=(6,4))
sns.heatmap(df[num].corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
# Q8
print('Average:',df.charges.mean())
print('Max:',df.charges.max())
print('Min:',df.charges.min())
print(df.groupby('smoker')['charges'].mean())
```

```python
# Q9
sns.boxplot(data=df,x='smoker',y='charges')
plt.show()
sns.boxplot(data=df,x='sex',y='charges')
plt.show()
```

- Average age: `df['age'].mean()`
- Average BMI: `df['bmi'].mean()`
- Smoking significantly increases insurance charges.
